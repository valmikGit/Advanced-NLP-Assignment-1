```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Input CSV (produced from your JSONL → CSV script)
INPUT_FILE = "mt_bench_human_judgements.csv"

def run_eda(input_file):
    # Load the CSV
    df = pd.read_csv(input_file)

    # Create a new column: "winner_type" → tie / not tie
    df["winner_type"] = df["winner"].apply(
        lambda x: "tie" if str(x).strip().lower() == "tie" else "not tie"
    )

    # Count distribution
    counts = df["winner_type"].value_counts()
    print("Winner type counts:\n", counts)

    # Plot
    plt.figure(figsize=(6, 5))
    sns.countplot(data=df, x="winner_type", palette="Set2")
    plt.title("Distribution of Winner Types")
    plt.xlabel("Winner Type")
    plt.ylabel("Count")

    # Show the plot interactively
    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    run_eda(INPUT_FILE)
```
